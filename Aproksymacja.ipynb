```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(mm,x,y):

    m = mm
    xsum = [np.sum(x**(j)) for j in range(2*m+1)]
    A = np.zeros([m+1,m+1])
    b = np.zeros(m+1)
    for j in range(0,m+1):
        for i in range(0,m+1):
            A[i][j] = xsum[(i+j)]
        b[j]=np.sum(y[a]*(x[a])**j for a in range(0,len(x)))
    Z = np.linalg.solve(A,b)
    return Z
        
    
```

```python

m=int(input("Podaj stopień dopasowanego wielomianu:"))


n=int(input("Podaj ilość punktów:"))
x =np.zeros(n,dtype=int)
y =np.zeros(n,dtype=int)
for i in range(n):
   temp=int(input("Podaj wartość x:"))
   x[i]=temp
   temp=int(input("Podaj wartość y:"))
   y[i]=temp




ZZ = func(m,x,y)
```

```python

plt.plot(x,y,"o", label="punkty")
step = np.arange(min(x)-1,max(x)+1,0.5)


res=[np.sum(ZZ[j]*(x**(j)) for j in range(0,m+1)) for x in step]

   
    
plt.plot(step, res,"--", label = "dopasowanie")
plt.legend()
plt.title("Dopasowanie")

```
